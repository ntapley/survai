```python
#for every e.g. fps frames we have 1 second 

#specify required metres travelled between each output frame (higher number means fewer frames)
dist = 10
#fraction of this distance covered per second (ASSUMING SPEED IS KPH)
frac_dps = (1000/3600)/dist 

vidcap = cv2.VideoCapture(fileloc+filename)
success,image = vidcap.read()
count = 0
sec = 0
#mpf = 0
dist_units = 1

while success:
  #a new file will be written for each meter
  cv2.imwrite("./output/frame%d.jpg" % dist_units, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
  
  #distance unit travelled per second  
  #speed is recorded at each second so we need to reference 
  #the correct second of the recording for speed (mps)
  dps= float(root[int(sec)][4].text[0:2])*frac_dps
  
  dpf=dps/fps #distance unit travelled per frame
  dist_units+=dpf #increment metres travelled
  sec += 1/fps #seconds since recording started
  print(count,dpf,dist_units,sec)
```
